#  Lists

```python
fruit_selection = ["apple", "banana", "cherry", "date", "elderberry"]
```

```python
# 1.Meyvələri əlifba sırasına görə sıralamaq üçün bir Python kodu yazın. Sıralanmış siyahını çap edin.
fruit_selection.sort()
print(fruit_selection)
```

```python
# 2. fruit_selection siyahısının sonuna "orange" meyvəsini əlavə edən bir Python kodu yazın. Daha sonra yenilənmiş siyahını çap edin.
fruit_selection.append("orange")
print(fruit_selection)
```

```python
# 3. fruit_selection siyahısının son elementini silin və onu çap edin.
silinen_element=fruit_selection.pop()
print("Silinen element:",silinen_element)
print(fruit_selection)
```

```python
numbers = [5, 4, 3, 1]
```

```python
# 4. Rəqəmlərin siyahısını kiçikdən böyüyə sıralayın və çatışmayan rəqəmləri əlavə edin.
numbers.sort()
numbers.insert(1,2)
print(numbers)
```

```python
text = "I-love-Data-analysis"
```

```python
# 5. Bu mətnin bir siyahıya çevrilməsi üçün Python kodu yazın. Siyahını "text_list" kimi saxlayın.
text_list=text.split("-")
print(text_list)
```

```python
# 6 Bu siyahını çap edin və siyahını yenidən mətnə çevirin.

print("Siyahı:", text_list)
yeniden_metn = " ".join(text_list)
print("Yenidən mətn:", yeniden_metn)
```

```python
# 7. "text" və "text_list" içərisindəki elementlərin sayını hesablayın və nəticələri müqayisə edərək hansının daha çox elementə malik olduğunu müəyyənləşdirin.
say_text=len(text)
say_list=len(text_list)
if say_text > say_list:
    print("Text daha çox elementə malikdir.")
elif say_list > say_text:
    print("Text_list daha çox elementə malikdir.")
else:
    print("Hər ikisinin element sayı bərabərdir.")

```

# Dictionaries

```python
product_dict = {
    "Laptop": ["Dell", 999.99],
    "Smartphone": ["Apple", 799.99],
    "Headphones": ["Sony", 129.95],
    "Tablet": ["Samsung", 349.99]
}
```

```python
# 1.product_dict lüğətinin açarlarını (keys) və dəyərlərini (values) çap edin.
print("acarlar:",product_dict.keys())
print("deyerler:",product_dict.values())
```

```python
# 2. "Smartphone": ["Apple", 799.99] yazısını "Smartphone": ["Samsung", 500] olaraq dəyişdirin.
product_dict["Smartphone"]=["Samsung", 500]
print(product_dict)
```

```python
# 3. "Smartphone" açarının dəyərini çap edin.
print(product_dict["Smartphone"])
```

# Tuple

```python
my_tuple = ("Auto", "Fahrrad", "Motorrad")
```

```python
# 1. Python-da my_tuple tuplenin ilk və son elementini çap edin.
print("Ilk element:", my_tuple[0])
print("Son element:", my_tuple[-1])
```

```python
# 2 Python-da my_tuple  tuplenin ilk elementini silin.
list_1 = list(my_tuple)
list_1.pop(0)
my_tuple = tuple(list_1)
print(my_tuple)
```

# Set

```python
# 1 "Auto", "Fahrrad" və "Motorrad" dəyərləri ilə bir Python seti yaradın.
my_set={"Auto", "Fahrrad" , "Motorrad"}
```

```python
# 2 Artıq yaradılmış setə "Auto" elementini əlavə edin.
my_set.add("Auto")
print(my_set)
```

```python
# 3 my_set-dəki elementlərin sayını hesablayın və my_set-i çap edin.

sayi = len(my_set)
print("Elementlərin sayı:", sayi)
print("Set:", my_set)
```
